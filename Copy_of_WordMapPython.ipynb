Import Dependecies

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd
import codecs
import requests
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
```

Import library to work with Images

```python
from PIL import Image
```

Import DataSet

```python
mydata_df=pd.read_csv("/content/Constitution Of India.csv")
```

```python
mydata_df.head()
```

```python
text_array=mydata_df["Articles"].values
```

```python
#convert it into string
text_string="".join(text_array)
```

```python
#replace it with empty string
text=text_string.replace("\n", "")
```

Import the image

```python
from IPython.display import display #it is for display images
from matplotlib.pyplot import imshow
IMG = Image.open(requests.get("https://www.pngarts.com/files/3/India-Map-Transparent-Images.png", stream=True).raw)
display(IMG)
```

```python
IMG_mask = Image.new("RGB", IMG.size, (255,255,255))

IMG_mask.paste(IMG,IMG)

IMG_mask = np.array(IMG_mask)
```

```python
stopwords = set(STOPWORDS)
stopwords.add("WHEN_CASE_STOPWORDS_NEED")

wordcloud = WordCloud(width = 1000,
                      height = 1500,
                      max_words = 500,
                      stopwords = stopwords,
                      mask=IMG_mask,

                      contour_width=7,contour_color='white', max_font_size = 200,random_state=40 ).generate(text)

fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
# display(wordcloud)
plt.axis('off')
```

```python
wordcloud = WordCloud(width = 1000,
                      height = 1500,
                      max_words = 500,
                      stopwords = stopwords,
                      mask=IMG_mask,
                      background_color = "orange",
                      contour_width=7,
                      contour_color='green',
                      max_font_size = 200,
                      
                      random_state=40
                     ).generate(text)

fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis('off')
```
